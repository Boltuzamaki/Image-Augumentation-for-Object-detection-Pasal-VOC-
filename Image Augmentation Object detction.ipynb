```python
import xml.etree.ElementTree as ET
import re
import numpy as np
from sklearn import preprocessing
def co_ordinate_extract(file):
    lists=[]
    tree = ET.parse(file)
    root = tree.getroot()
    xmlstr = ET.tostring(root, encoding='utf8', method='xml')
    xmlstr = xmlstr.decode('utf-8')
    folder = re.search('<folder>(.*)</folder>', xmlstr)
    file = re.search('<filename>(.*)</filename>', xmlstr)
    label = re.findall('<name>(.*)</name>', xmlstr)
    ymax = re.findall('<ymax>(.*)</ymax>', xmlstr)
    xmax = re.findall('<xmax>(.*)</xmax>', xmlstr)
    xmin = re.findall('<xmin>(.*)</xmin>', xmlstr)
    ymin = re.findall('<ymin>(.*)</ymin>', xmlstr)
    ywidth = re.findall('<width>(.*)</width>',xmlstr)
    yheight = re.findall('<height>(.*)</height>',xmlstr)
    le = preprocessing.LabelEncoder()
    le.fit(label)
    class_f = le.transform(label)
    for i in range(0, len(label)):
        lists.append([xmin[i],ymin[i],xmax[i],ymax[i]])
    lists = np.array(lists)
    class_f = class_f.reshape(len(label),1)
    final = np.append(lists, class_f , axis=1)
    final = final.astype('float64')
    return final,yheight,ywidth,le
```

```python
from pascal_voc_writer import Writer
def scale_all(img_folder,xml_folder,savepath,scale=0.2):
    images_lis = os.listdir(img_folder)
    xml_lis = os.listdir(xml_folder)
    for xml in xml_lis:
        image = xml.split('.')[0]+str('.jpg')
        os.chdir(xml_folder)
        final,yheight,ywidth,le = co_ordinate_extract(xml)
        xml_name = xml.split('.')[0]+'_scale_'+'.xml'
        img_name = xml.split('.')[0]+'_scale_'+'.jpg'
        os.chdir(img_folder)
        img = cv2.imread(image)[:,:,::-1]
        transforms = RandomScale(scale)
        img, bboxes = transforms(img, final)
        final = final.astype('int64')
        # Creating xml
        location_image = savepath+"//"+str(img_name)
        writer = Writer(location_image, ywidth, yheight)
        for i in range(0, final.shape[0]):
            writer.addObject(le.inverse_transform([final[i][4]])[0], final[i][0], final[i][1], final[i][2], final[i][3])
        os.chdir(savepath)
        location_xml = savepath+"//"+str(xml_name)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        cv2.imwrite(img_name,img)
        writer.save(xml_name)
```
